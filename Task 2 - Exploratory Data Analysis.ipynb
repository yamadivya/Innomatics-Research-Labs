# Feature Engineering

we will select the features that are important 

```python
features = df[['Gender', '10percentage', '12percentage', 'Specialization', 'CollegeTier', 'GraduationYear', 'English', 'Logical', 'Quant', 'Domain', 'ComputerProgramming', 'ElectronicsAndSemicon', 'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg', 'conscientiousness', 'agreeableness', 'extraversion', 'nueroticism', 'openess_to_experience']]
features.head()
```

## Creating dummy variables to categorical columns

```python
features_one_hot = pd.get_dummies(features, columns=['Specialization'])
features_one_hot.head()
```

## Converting to Binary for Gender feature

```python
binary = {'m': 1, 'f': 0}
df['Gender'] = df['Gender'].map(binary)
```
